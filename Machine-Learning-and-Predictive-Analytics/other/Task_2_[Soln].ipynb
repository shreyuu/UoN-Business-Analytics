```python
import pandas
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.dummy import DummyRegressor
from sklearn.model_selection import cross_val_score, KFold

orig_data = pandas.read_csv('http://www.cs.nott.ac.uk/~pszgss/teaching/ML/Prac1/winequality-white_train.csv', header = 0, skiprows = [1,2])
new_data = pandas.read_csv('http://www.cs.nott.ac.uk/~pszgss/teaching/ML/Prac1/winequality-white_test.csv', header = 0, skiprows = [1,2])

output_feature_orig = orig_data.quality
input_features_orig = orig_data.drop('quality',axis = 1) # 0 rows, 1 cols

output_feature_new = new_data.quality
input_features_new = new_data.drop('quality',axis = 1) # 0 rows, 1 cols


rf = RandomForestRegressor()
ml = DummyRegressor(strategy='mean')

folds = KFold(n_splits=10, shuffle=True, random_state=0)
scores = {}
scores['Mean Learner, MAE'] = cross_val_score(ml, input_features_orig, output_feature_orig, scoring = 'neg_mean_absolute_error', cv=folds)
scores['Mean Learner, MSE'] = cross_val_score(ml, input_features_orig, output_feature_orig, scoring = 'neg_mean_squared_error', cv=folds)
scores['Random Forest, MAE'] = cross_val_score(rf, input_features_orig, output_feature_orig, scoring = 'neg_mean_absolute_error', cv=folds)
scores['Random Forest, MSE'] = cross_val_score(rf, input_features_orig, output_feature_orig, scoring = 'neg_mean_squared_error', cv=folds)
```

```python
# Print the results
for k in sorted(scores):
    print('{0:31}: {1:5.2f}'.format(k, -np.mean(scores[k])) )
```

```python
rf.fit(input_features_orig,output_feature_orig)
ml.fit(input_features_orig,output_feature_orig)

rf_pred = rf.predict(input_features_new)
ml_pred = ml.predict(input_features_new)


from sklearn.metrics import mean_absolute_error, mean_squared_error
ml_MAE = mean_absolute_error(output_feature_new, ml_pred)
rf_MAE = mean_absolute_error(output_feature_new, rf_pred)

ml_MSE = mean_squared_error(output_feature_new, ml_pred)
rf_MSE = mean_squared_error(output_feature_new, rf_pred)

print('ML: {:.2f} (MAE) {:.2f} (MSE)'.format(ml_MAE,ml_MSE))
print('RF: {:.2f} (MAE) {:.2f} (MSE)'.format(rf_MAE,rf_MSE))
```
