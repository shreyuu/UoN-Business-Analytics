```python
import pandas
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.model_selection import cross_val_score, StratifiedKFold

data = pandas.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/00267/data_banknote_authentication.txt', header = None)
data.columns = ['varval','skewval','curtval','entval','classval']
data.head(5)
```

```python
# Replacing the Select Columns Widget
X = data.drop('classval',axis = 1)
y = data.classval

# Learners
rf = RandomForestClassifier()
lg = LogisticRegression()
ml = DummyClassifier()

# Replacing the Test & Score Widget
folds = StratifiedKFold(n_splits=10, shuffle=True, random_state=0)
scores = {}
scores['Logistic Regression, CA'] = cross_val_score(lg, X, y, scoring = 'accuracy', cv=folds)
scores['Logistic Regression, Precision'] = cross_val_score(lg, X, y, scoring = 'precision', cv=folds)
scores['Logistic Regression, Recall'] = cross_val_score(lg, X, y, scoring = 'recall', cv=folds)
scores['Random Forest, CA'] = cross_val_score(rf, X, y, scoring = 'accuracy', cv=folds)
scores['Random Forest, Precision'] = cross_val_score(rf, X, y, scoring = 'precision', cv=folds)
scores['Random Forest, Recall'] = cross_val_score(rf, X, y, scoring = 'recall', cv=folds)
scores['Majority Learning, CA'] = cross_val_score(ml, X, y, scoring = 'accuracy', cv=folds)
scores['Majority Learning'] = cross_val_score(ml, X, y, scoring = 'precision', cv=folds)
scores['Majority Learning'] = cross_val_score(ml, X, y, scoring = 'recall', cv=folds)

# Alternatively we can just specify cv = 10 and have the cross_val_score automatically use a StratifiedKFold object
# scores = {}
# scores['Logistic Regression, CA'] = cross_val_score(lg, X, y, scoring = 'accuracy', cv=10)
# scores['Logistic Regression, Precision'] = cross_val_score(lg, X, y, scoring = 'precision', cv=10)
# scores['Logistic Regression, Recall'] = cross_val_score(lg, X, y, scoring = 'recall', cv=10)
# scores['Random Forest, CA'] = cross_val_score(rf, X, y, scoring = 'accuracy', cv=10)
# scores['Random Forest, Precision'] = cross_val_score(rf, X, y, scoring = 'precision', cv=10)
# scores['Random Forest, Recall'] = cross_val_score(rf, X, y, scoring = 'recall', cv=10)
# scores['Majority Learning, CA'] = cross_val_score(ml, X, y, scoring = 'accuracy', cv=10)
# scores['Majority Learning'] = cross_val_score(ml, X, y, scoring = 'precision', cv=10)
# scores['Majority Learning'] = cross_val_score(ml, X, y, scoring = 'recall', cv=10)

# Alternatively you could have used cross_validate(...) rather than cross_val_score(....)

# Print the results
for k in sorted(scores):
    print('{0:31}: {1:5.2f}%'.format(k, np.mean(scores[k])*100) )
```
