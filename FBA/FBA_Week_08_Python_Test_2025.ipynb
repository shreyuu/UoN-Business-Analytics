### Q5.1: Defining a basic function (2 marks)

Write a function `invert` that takes a number as an argument and returns its inverse.

For example, if the input is 5, the function returns 1/5.

```python
#enter your answer here
# def invert_1(n):
#     m=len(n)-1
#     p=""
#     while m>=0:
#         p+=n[m]
#         m-=1
#     print((p))

def invert_2(n):
    rev=0
    while(n>0):
        dig=n%10
        rev=rev*10+dig 
        n=n//10 
    print("Reverse of the number: ",rev)

invert_2(5)
```

### Q5.2: Manipulating a list with a function (4 marks)

image.png

```python
numbers = [1, 3, 2, 10, 8, 6, 9, 22, 23]
# You may try different values to test your code

#enter your answer here
def even_numbers(n):
    n.sort()
    res = [val for val in n if val % 2 == 0]
    return res

#test using the following function calls
print(even_numbers(numbers)) # Should print [2, 6, 8, 10, 22]
```
